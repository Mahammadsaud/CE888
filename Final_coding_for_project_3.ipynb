```python
#Last outcome category is not replaced by 0, So we replace it with zero event
new_data = data.fillna({
        'Longitude': 0,
        'Latitude': 0,
        'Crime type': 0,
        'Last outcome category': 'No Outcome'
            })
new_data
```
